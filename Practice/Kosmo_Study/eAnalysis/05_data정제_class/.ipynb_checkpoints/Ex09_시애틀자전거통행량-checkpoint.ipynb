```python
# 1. 데이타 받아오기
# https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD
# 시애틀 프리몬트 다리에 동쪽과 서쪽 보도에 유도 센서를 장착하여 자전수 수를 집계하였다

import pandas as pd

data=pd.read_csv('./data/Fremont_Bridge.csv', index_col='Date', parse_dates=True)
data.head()  # Date 컬럼으로 인덱스를 하고 문자열에서 Datetime 오브젝트로 설정

# 컬럼명을 다시 지정하고 Total 컬럼 추가
data.columns=['East','West']
data['Total']=data['East']+data['West']
data

# 데이타 확인
data.isnull().any()
data.dropna()
data.describe()
```

```python
# 2. 데이타 시각화하기
import matplotlib.pyplot as plt
import seaborn as sb

data.plot()
plt.ylabel('Hourly Bicycle Count')
```

```python
"""(1) 리샘플링 - 주 단위로
      너무 많은 데이타(시간별 표본)이 밀집되어 이해하기 어렵다"""
weekly = data.resample('W').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Weekly Bicycle Count')

# [결과] 예상대로 여름이 겨울보다 자전거 사용 횟수가 높다
#        2017년도부터 서쪽다리 이용률이 동쪽보다 높다
#       ( 서쪽은 시애틀 도심으로 들어올 때 사용하고 동쪽은 시애틀 도심을 벗어날 때 사용한단다 )
```

```python
 """(2) 리샘플링 - 일 단위로 추출하고 
    데이타의 50일 롤링 평균을 구해 평균값이 중심에 오도록함"""
daily = data.resample("D").sum()
dailysum = daily.rolling(50, center=True).sum()
dailysum.plot(style=[':','--','-'])
plt.ylabel('mean hourly count ')

# 가우스 윈도우함수로 부드럽게 표현 - 윈도우 폭을 50일로 하고 윈도우 내 가우스 폭ㅇ르 10일로 지정
dailysum = daily.rolling(50, center=True, win_type='gaussian').sum(std=10)
dailysum.plot(style=[':','--','-'])
```

```python
""" (3) 데이타 분석하기
        하루의 시간대를 기준으로 평균 통행량
"""
import numpy as np

by_time = data.groupby(data.index.time).mean()
by_time
hourly_ticks = 4*60*60*np.arange(6) # 4시간 단위 초?
hourly_ticks

by_time.plot(xticks=hourly_ticks, style=[':','--','-'])
# [결론] 아침 8시와 저녁 5시경에 사용량이 많다는 것은 출퇴근시간임을 추측할 수 있다
```

```python
""" 요일에 따른 통행량 """
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']  
by_weekday
by_weekday.plot(style=[':','--','-']) # 인덱스 출력안되는데?
# [결과] 주중 통행량은 주말 통행량의 2배 이상이다.
```

```python
""" 주중과 주말에 시간대별 추이 """
# print(data.index.weekday)
weekend = np.where(data.index.weekday<5,'WEEKDAY', 'WEEKEND')
# print(weekend)  # 숫자 0~6를 'WEEKDAY'와 'WEEKEND'로 지정

by_time = data.groupby([weekend, data.index.time]).mean()
# by_time

by_time.loc['WEEKDAY'].plot(title='Weekdays', xticks=hourly_ticks, style=[':','--','-'])
by_time.loc['WEEKEND'].plot(title='Weekends', xticks=hourly_ticks, style=[':','--','-'])

# 다중 서브플랫으로 표현하려면
# fig, ax = plt.subplots(1, 2, figsize=(14,5))
# by_time.loc['WEEKDAY'].plot(ax=ax[0],title='Weekdays', xticks=hourly_ticks, style=[':','--','-'])
# by_time.loc['WEEKEND'].plot(ax=ax[1],title='Weekends', xticks=hourly_ticks, style=[':','--','-'])

# [결과] 주중에는 출퇴근 시간에 이용량이 많고 주말에는 오후에 이용을 많이 한다.
```
