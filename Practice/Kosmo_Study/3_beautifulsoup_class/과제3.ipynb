 2) 이미지 다운

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
from urllib import request
from selenium.webdriver.common.keys import Keys
import time
```

```python
book_list=[]
sell_book=0
price_sum=0
for i in range(1,6730):
    try:
        url = 'https://wikidocs.net/book/'+str(i)
        site = request.urlopen(url)
        content = site.read()
        soup = BeautifulSoup(content,'html.parser')

        title = soup.select_one('#load_content > h1').text
        title =title.replace("  ","")
        title = title.strip()
        print(title)

        price = soup.select_one('#load_content ').text
        row = soup.select_one('.book-info').select('.row')
        price =row[3].select_one('strong').text
        price = price.replace(",","")
        if int(price) < 5000:
            print("가격이 매겨지지않은 책입니다.")
            price="0"
        price_sum += int(price) 
        print(price)
        
        p_list = soup.select('.page-content > p')
        content =""
        for p in p_list:
            content += p.text
        print(content)

        img = soup.select_one('#load_content > div.book-info.row.clearfix > div.book-image-box.pull-left > img')


        img_path = url+img.attrs['src']
        print(img_path)
        request.urlretrieve(img_path,'img2/' + title +'.jpg')

        sell_book += 1
        
        print("보유한 책들의 가치 :",price_sum)
        print("총 몇 권",sell_book)
        print("-"*20)
        book = [title,price,content]
        book_list.append(book)
        
    except:
        print("삭제된 책")
        print("-"*20)
```
