```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# 교보문고 > '파이썬' 검색 > 국내도서
html = urlopen("http://www.kyobobook.co.kr/search/SearchKorbookMain.jsp?vPstrCategory=KOR&vPstrKeyWord=python&vPplace=top")
```

## 1) 드라이버 선언 & url 적용

```python
driver = webdriver.Chrome("C:/tool/chromedriver.exe")
driver.get("http://www.kyobobook.co.kr/search/SearchKorbookMain.jsp?vPstrCategory=KOR&vPstrKeyWord=python&vPplace=top")
driver.implicitly_wait(3)
```

```python
title_list = []
paging = driver.find_element_by_id('paging')
paging = paging.find_elements_by_tag_name('li')
total_page = driver.find_element_by_id('totalpage').text
total_page = int(total_page)


print(paging[0].text)

total = total_page/10
rest = total_page%10
for x in range(1,int(total)+1):
    driver.get("https://search.kyobobook.co.kr/web/search?vPstrKeyWord=python&vPstrTab=PRODUCT&searchPcondition="+str(x)+"&currentPage=6#container")
    driver.implicitly_wait(3)
    for i in range(0,len(paging)):
        paging = driver.find_element_by_id('paging')
        paging = paging.find_elements_by_tag_name('li')
        paging[i].click()

        time.sleep(0.5)
        tbody = driver.find_element_by_id('search_list')
        tr_list = tbody.find_elements_by_tag_name('tr')
        for tr in tr_list:
            a = tr.find_element_by_class_name('title').find_element_by_tag_name('a')
            title_list.append(a.text)
            print(a.text)

        time.sleep(0.25)
    driver.find_elements_by_class_name('btn_next')[0].click



```
