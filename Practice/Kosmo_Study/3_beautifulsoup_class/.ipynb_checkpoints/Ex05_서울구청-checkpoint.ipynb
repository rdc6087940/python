```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
from urllib import request
from selenium.webdriver.common.keys import Keys
import time
```

## 1. 웹 문서 가져오기
- https://www.seoul.go.kr/seoul/autonomy.do

```python
url = 'https://www.seoul.go.kr/seoul/autonomy.do'
site = request.urlopen(url)
content = site.read()
print(content)
```

## 2. 웹문서에 추출한 데이터를 리스트에 저장

```python
soup = BeautifulSoup(content,'html.parser')
li_list = soup.select('.call-list > ul > li')
print(li_list[0].text.split()[0])

구청이름 = []
구청주소 = []
구청전화 = []
구청홈페이지 = []
for li in li_list:
    lists = li.select('li')
    구청이름.append(li.text.split()[0])
    구청주소.append(lists[0].text)
    구청전화.append(lists[1].text)
    구청홈페이지.append(lists[2].text)

```
