```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
from urllib import request
from selenium.webdriver.common.keys import Keys
import time

url = 'https://search.kyobobook.co.kr/web/search?vPstrKeyWord=python&vPstrTab=PRODUCT&searchPcondition=3&currentPage=3#container'
site = request.urlopen(url)
page = site.read()

soup = BeautifulSoup(page, 'html.parser')
cover_list = soup.select('#search_list .cover img')
img_list = []
for cover in cover_list:
    img_list.append(cover.attrs['src'])
    img_name = cover.attrs['alt'].strip().replace('/',"")
    img_path = cover.attrs['src']
    request.urlretrieve(img_path,'data/' + img_name +'.jpg')
       

print(len(img_list))
```
